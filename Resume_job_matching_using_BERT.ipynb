```python
# Step 2: Preprocess the Resume Texts

def clean_resume(text):
    text = re.sub('http\S+|www\S+', '', text) # Remove URLs
    text = re.sub('[^A-Za-z0-9\s]+', '', text) # Remove special characters
    text = re.sub('\s+', ' ', text).strip() # Normalize whitespace
    return text

data['cleaned_resume'] = data['Resume'].apply(clean_resume)
```

```python
data.head()
```

```python
# Step 3: Encode Resume Texts using Sentence-BERT
from sentence_transformers import SentenceTransformer

sbert_model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Generate embeddings
resume_embeddings = sbert_model.encode(data['cleaned_resume'].tolist(), show_progress_bar=True)
```

```python
# Step 4: Load and Preprocess Job Descriptions
# Example JDs (can be replaced with real JD data)
jds = [
    "Looking for a data analyst with Python, SQL, and data visualization experience.",
    "Seeking a machine learning engineer with experience in deep learning and PyTorch.",
    "We need a web developer skilled in HTML, CSS, JavaScript, and React."
]

jd_embeddings = sbert_model.encode(jds)
```

```python
# Step 5: Match Resumes to Job Descriptions (Cosine Similarity)
from sklearn.metrics.pairwise import cosine_similarity

for i, resume_vec in enumerate(resume_embeddings):
    scores = cosine_similarity([resume_vec], jd_embeddings)[0]
    ranked_jobs = sorted(list(enumerate(scores)), key=lambda x: x[1], reverse=True)
    print(f"\nTop matches for Resume {i+1} ({data['Category'].iloc[i]}):")
    for idx, score in ranked_jobs:
        print(f"  JD {idx+1}: Score = {score:.4f} | {jds[idx]}")
```
